```python
import pdfplumber

# Define a function to extract tables from the PDF
def extract_tables_from_pdf(pdf_path):
    all_tables = []

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            # Extract tables from the page
            tables = page.extract_tables()

            # If tables are found, process them
            if tables:
                for table in tables:
                    all_tables.append(table)

    return all_tables

# Extract tables from the provided PDF
pdf_path = "/content/Documents/am8100_ba_en.pdf"
tables = extract_tables_from_pdf(pdf_path)

# Print all the rows of each table
for table in tables:
    print("Table")  # Title for each table
    for row in table:
        print(row)
    print()  # One space line between tables

# Save all the rows of each table to a .txt file named "tablescard.txt"
with open("tablescard.txt", "w") as txt_file:
    for table in tables:
        txt_file.write("Table\n")  # Title for each table
        for row in table:
            # Write the row as a list representation to the file
            txt_file.write(str(row))
            txt_file.write("\n")  # One new line for spacing between rows
        txt_file.write("\n")  # One new line for spacing between tables
```
